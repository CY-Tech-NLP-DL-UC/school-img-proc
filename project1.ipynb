### Task 1 : binary images

**1.1** Create an image that is composed of a 100x100 pixels black square and a 50x50 pixels white square in the middle using 2 ways  :

* Create a big black square 100x100 pixels then change the right pixels to white.
* Create a big black square 100x100 pixels then a 50x50 pixels white square and paste it at the right spot.

```python
# first method
img = np.zeros((100, 100)) # 0 equal to black pixels
img[25:75, 25:75] = 1 # 1 equal to white pixels
plt.imshow(img, cmap=plt.cm.binary_r)
```

```python
# second method
img1 = np.zeros((100,100))
img2 = np.ones((50,50))
img1[25:75, 25:75] = img2
plt.imshow(img1, cmap=plt.cm.binary_r)
```

*Advanced question* : turn this into parameters to modify background size, foreground size, and its position

```python
def create_square(size_bg, size_sq, pos):
    img = np.zeros(size_bg)
    img[pos[0]:pos[0]+size_sq[0], pos[1]:pos[1]+size_sq[1]] = np.ones(size_sq)
    plt.imshow(img, cmap=plt.cm.binary_r)
```

```python
create_square((80,80),(10,10),(20,60))
```

### Task 2 :

**2.1** Create a binary image of size 3x3 pixels with alternated values of pixels (1 and 0)

```python
def alternated_square(size_sq):
    img = np.zeros(size_sq)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if (i+j+2)%2:
                img[i,j] = 1
    return img
```

```python
img = alternated_square((3,3))
plt.imshow(img, cmap=plt.cm.binary_r)
```

**2.2** Vary images size to 4x4 pixels and 5x5 pixels

```python
img = alternated_square((4,4))
plt.imshow(img, cmap=plt.cm.binary_r)
```

```python
img = alternated_square((5,5))
plt.imshow(img, cmap=plt.cm.binary_r)
```

### Task 3 : Create and show an image of intensity levels of 4x7 pixels with vertical lines with layers progressively going from black to white

```python
img = np.zeros((4, 7))
for i in range(0, img.shape[1]):
    img[:, i] = i*(1/6)
plt.imshow(img, cmap=plt.cm.binary_r)
```

### Task 5 : Write a program to an image of size N x N with progressive layers of gray

```python
def layers_gray(size):
    img = np.zeros(size)
    for i in range(img.shape[0]):
        img[i] = i/(img.shape[1] - 1)
    return img
```

```python
size =(100,100)
plt.imshow(layers_gray(size), cmap=plt.cm.binary_r)
```

### Task 7 : indexed color images (RGB)

**7.1** Create matrix of color map to obtain the image of color :

|   Gray        |   Blue-green   |   Red   |

|   Blue-Green   |   Red       |   Blue-Green   |

|   Red       |   Blue-Green   |   Gray    |

```python
import matplotlib as mlp
map = mlp.colors.ListedColormap(
    [
        (1, 0, 0), # Red 0
        (0, 0.6, 1), # Blue-green 1
        (0.5, 0.5, 0.5) # Gray 2
    ]
)
img = np.array([
    [2, 1, 0],
    [1, 0, 1],
    [0, 1, 2]
])
plt.imshow(img, map)
```
